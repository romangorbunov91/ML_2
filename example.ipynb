Найти один из корней уравнения f(x)=x^3−3x^2+2x−5=0 в окрестности точки x=2 с помощью градиентного спуска. Используйте метод минимизации функции потерь L(x)=(f(x))^2.

```python
import numpy as np

def f(x):
    return x**3 - 3*x**2 + 2*x - 5

def f_prime(x):
    return 3*x**2 - 6*x + 2

def gradient_descent(initial_x, learning_rate, max_iterations, tolerance):
    x = initial_x
    for i in range(max_iterations):
        loss = f(x)**2
        grad = 2 * f(x) * f_prime(x)  # Градиент L(x)
        x_new = x - learning_rate * grad
        
        # Проверка условия остановки
        if abs(x_new - x) < tolerance:
            break
        x = x_new
        
        # Вывод информации о шаге (опционально)
        if i % 100 == 0:
            print(f"Iteration {i}: x = {x:.5f}, f(x) = {f(x):.5f}")
    
    return x

# Параметры алгоритма
initial_x = 2.0  # Начальная точка в окрестности x=2
learning_rate = 0.01  # Скорость обучения
max_iterations = 1000  # Максимальное число итераций
tolerance = 1e-6  # Условие остановки

# Запуск градиентного спуска
root = gradient_descent(initial_x, learning_rate, max_iterations, tolerance)

# Вывод результата
print("\nНайденный корень: x = {:.5f}".format(root))
print("f({:.5f}) = {:.5f}".format(root, f(root)))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 3*x**2 + 2*x - 5

# Создаем диапазон значений x от 0 до 5
x = np.linspace(0, 5, 400)
y = f(x)

# Строим график
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$f(x) = x^3 - 3x^2 + 2x - 5$')
plt.axhline(0, color='black', linewidth=0.5)  # Горизонтальная ось
plt.grid(True, linestyle='--', alpha=0.7)

# Отмечаем примерное положение корня из предыдущего решения
root_x = root  # Значение из градиентного спуска
plt.scatter(root_x, f(root_x), color='red', label=f'Корень (~{root_x:.3f})')

plt.title('График функции $f(x) = x^3 - 3x^2 + 2x - 5$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

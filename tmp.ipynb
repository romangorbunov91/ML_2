```python
import numpy as np

def f(x):
    return x**3 - 3*x**2 + 2*x - 5

def f_deriv(x):
    return 3*x**2 - 6*x + 2

def gradient_descent(initial_x, lr, max_iterations, tolerance):
    x = initial_x
    for i in range(max_iterations):
        loss = f(x)**2
        grad = 2 * f(x) * f_deriv(x)  # Градиент L(x)
        x_new = x - lr * grad
        
        # Проверка условия остановки
        if abs(x_new - x) < tolerance:
            x = x_new
            break
        x = x_new
        
        # Вывод информации о шаге (опционально)
        if i % 100 == 0:
            print(f"Iteration {i}: x = {x:.5f}, f(x) = {f(x):.5f}")
    
    return x

# Параметры алгоритма
initial_x = 2.0  # Начальная точка в окрестности x=2
lr = 0.01  # Скорость обучения
max_iterations = 1000  # Максимальное число итераций
tolerance = 1e-6  # Условие остановки

# Запуск градиентного спуска
root = gradient_descent(initial_x, lr, max_iterations, tolerance)

# Вывод результата
print("\nНайденный корень: x = {:.5f}".format(root))
print("f({:.5f}) = {:.5f}".format(root, f(root)))
```
